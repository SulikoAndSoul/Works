# 项目：评估和清理英国电商公司销售数据

## 分析目标

此数据分析的目的是，根据市场销售数据，挖掘畅销产品，以便制定更有效的市场策略来提升营收。

本实战项目的目的在于练习评估数据干净和整洁度，并且基于评估结果，对数据进行清洗，从而得到可供下一步分析的数据。

## 简介

原始数据集记录了一家英国在线零售公司在2010年12月1日至2011年12月9日期间的所有交易情况，涵盖了该公司在全球不同国家和地区的业务数据。该公司主要销售覆盖各个场景的礼品，包括但不限于生日礼品、结婚纪念品、圣诞礼品等等。该公司的客户群体主要包括批发商和个人消费者，其中批发商占据了相当大的比例。

数据每列的含义如下：
- `InvoiceNo`: 发票号码。6位数，作为交易的唯一标识符。如果这个代码以字母“c”开头，表示这笔交易被取消。
- `StockCode`: 产品代码。5位数，作为产品的唯一标识符。
- `Description`: 产品名称。
- `Quantity`: 产品在交易中的数量。
- `InvoiceDate`: 发票日期和时间。交易发生的日期和时间。
- `UnitPrice`: 单价。价格单位为英镑（£）。
- `CustomerID`: 客户编号。5位数，作为客户的唯一标识符。
- `Country`: 国家名称。客户所居住的国家的名称。

## 读取数据

```python
import pandas as pd
```

```python
orignal_data = pd.read_csv("./e_commerce.csv")
```

## 评估数据

#### 评估数据的结构性混乱

###### 我们可以通过样本提取查看数据是否遵从三原则

```python
orignal_data.sample(10)
```

###### 可以看得出来，“每列都对应一个变量，每行一个观察值，每单元格对应一个值”，该数据并不存在结构性问题

#### 评估数据的内容性混乱

##### 评估缺失值

```python
orignal_data.info()
```

###### 可以看出，在`541909`个数据中，`Description`与`CustomerID`的数据出现了缺失，需要进行提取具体查看是否会对畅销产品的查找造成影响
###### 而且`InvoiceDate`应为DataTime，`CustomerID`应为字符串

```python
orignal_data[orignal_data["Description"].isnull()]
```

###### 这些抽样数据中呈现出的特征都是`UnitPrice`为零，即便具不具备`Description`都无法呈现出畅销产品应具备的特征。
###### 但是否所有缺失`Description`的数据都是`UnitPrice`为零呢？这需要进一步的提取。

```python
orignal_data[(orignal_data["Description"].isnull()) & (orignal_data["UnitPrice"] != 0)]
```

###### 上述的零数据表明，所有缺失`Description`的数据都是`UnitPrice`为零。因此缺失`Description`数据可以直接清洗

###### 接下来查看`CustomerID`

###### 虽然`CustomerID`存在缺失，但`CustomerID`与`Country`指代的是同一事物，且`Country`比`CustomerID`更加直观，因此Series`CustomerID`可以直接删除

##### 评估重复值

```python
orignal_data[orignal_data.duplicated()]
```

###### 重复值多出现在`CustomerID`与`Country`中，这种重复值并不会影响畅销产品的查找，因此可以不予处理
###### 即便有`InvoiceNo`、`StockCode`、`Description`的数据出现重复，但其可以理解为多个订单，因此可以不予处理

##### 评估不一致值

###### 此评估旨在找出不同值但表示同一现实对象的值
###### 在每个列名中，唯独`Country`很有可能出现不同值但表示同一现实对象的值，如"UK""United Kingdom"都表示英国，因此要予以评估

###### 可以先通过对每个出现的值的个数进行总结，在查找是否出现不同值但表示同一现实对象的值

```python
orignal_data["Country"].value_counts()
```

###### 这里可以发现`United Kingdom`、`UK`、`U.K.`都是指代英国，`USA`、`United States`都是指代美国，这无疑需要在清洗数据中进行统一

##### 评估无效或错误值

```python
orignal_data.describe()
```

###### 令人惊讶的是，`Quantity`数量与`UnitPrice`单价竟然出现了负数，我们应该进行提取然后分析是否会对畅销产品的查找造成问题

###### 首先查看`Quantity`为负数的数据

```python
orignal_data[orignal_data["Quantity"] < 0]
```

###### 可以看到，抽样出来`Quantity`为负数的数据，都具有`InvoiceNo`含"C"的情况（即被取消订单），那么是否所有`Quantity`为负数的数据都是被取消订单呢？

```python
orignal_data[(orignal_data["Quantity"] < 0) & (orignal_data["InvoiceNo"].str[0] !="C")]
```

###### 事实并非如此，所有`Quantity为`负数的数据只有部分是被取消订单，但同时又可以发现，这些不畏被取消订单的数据：`UnitPrice`为零，这又不会影响畅销产品的查找。
###### 所以我们可以增加条件，看是否可以排除`Quantity`为负数的所有数据

```python
orignal_data[(orignal_data["Quantity"] < 0) & (orignal_data["InvoiceNo"].str[0] !="C") & (orignal_data["UnitPrice"] != 0)]
```

###### 这下没有了！也就是说，`Quantity`为负数的数据，不是被取消订单，就是单价为零，无法造成对畅销产品的影响，因此可以被清洗掉。

###### 我们再继续查看`UnitPrice`为负数的数据

```python
orignal_data[orignal_data["UnitPrice"] < 0]
```

###### 此为坏账数据，不会对畅销产品的查找造成影响，可以大胆清洗

## 清理数据

#### 根据数据的评估，姑且可以做出以下清洗步骤：
###### 将`InvoiceDate`变为datatime，`CustomerID`变为字符串
###### 直接清洗缺失`Description`的数据
###### 将不一致值进行统一
###### 清洗`Quantity`、`UnitPrice`为负数的数据

```python
cleaned_data = orignal_data.copy()
```

##### 将`InvoiceDate`变为datatime，`CustomerID`变为字符串

```python
cleaned_data["InvoiceDate"] = pd.to_datetime(cleaned_data["InvoiceDate"])
cleaned_data["InvoiceDate"]
```

```python
cleaned_data["CustomerID"].astype(str)
```

```python
cleaned_data["CustomerID"] = cleaned_data["CustomerID"].slice(0,-2)
```

##### 直接清洗缺失`Description`的数据

```python
cleaned_data.dropna(subset=["Description"],inplace=True)
cleaned_data["Description"].isnull().sum()
```

##### 将不一致值进行统一

```python
cleaned_data["Country"] = cleaned_data["Country"].replace(["UK","U.K."],"United Kingdom")
cleaned_data["Country"] = cleaned_data["Country"].replace("USA","United States")
```

```python
cleaned_data["Country"].value_counts()
```

##### 清洗`Quantity`、`UnitPrice`为负数的数据

```python
cleaned_data = cleaned_data[cleaned_data["Quantity"] >= 0]
cleaned_data[cleaned_data["Quantity"] < 0]
```

```python
cleaned_data = cleaned_data[cleaned_data["UnitPrice"] >= 0]
cleaned_data[cleaned_data["UnitPrice"] < 0]
```

## 保存清理后的数据

```python
cleaned_data.info()
```

```python
cleaned_data.to_csv("e_commerce_cleaned.csv",index=False)
```
