```python
import pandas as pd
```

```python
orignal_data = pd.read_csv('Data.csv')
```

```python
cleaned_data = orignal_data.drop(['Countries','Years'],axis=1)
cleaned_data
```

```python
import statsmodels.api as sm
```

```python
lr_data = cleaned_data.copy()
lr_data
```

```python
y = lr_data['Y']
x = lr_data[['X1','X2']]
```

```python
x.corr().abs() > 0.8
```

```python
x = sm.add_constant(x)
```

```python
lr_model = sm.OLS(y,x).fit()
lr_model
```

```python
lr_model.summary()
```

```python
import matplotlib.pyplot as plt
```

```python
data_to_predict = pd.read_excel('Data_to_predict.xlsx')
data_to_predict.drop(["Countries","Years"],axis=1,inplace=True)
```

```python
data_to_predict
```

```python
x_predict = data_to_predict[['X1','X2']]
x_predict = sm.add_constant(x_predict)
```

```python
data_predict = lr_model.predict(x_predict)
```
